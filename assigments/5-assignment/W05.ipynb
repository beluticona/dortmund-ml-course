## Data Preparation

```python
#target variable y
y_training = train_set.iloc[:,-1]
y_testing = test_set.iloc[:,-1]
```

```python
#input variables x
X_training= train_set.iloc[:,:9]
X_testing = test_set.iloc[:,:9]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

def linear_regression_r2_score(X_train, X_test, y_train, y_test):
  clf_lr = LinearRegression(n_jobs = -1)
  clf_lr.fit(X_train,y_train)
  y_pred = clf_lr.predict(X_test)
  y_pred_train = clf_lr.predict(X_train)
  return {'train':r2_score(y_train,y_pred_train), 'test':r2_score(y_test,y_pred)}  
```

Ex 1: linear regression model to the training data using all input variables.

```python
linear_regression_r2_score(X_training, X_testing, y_training, y_testing)
```

Ex2: Linear regression model to the training data using the first 6 input variables

```python
linear_regression_r2_score(X_training.iloc[:,:6], X_testing.iloc[:,:6], y_training, y_testing)
```

Ex3: Linear regression model to the training data using the first 7 input variables

```python
linear_regression_r2_score(X_training.iloc[:,:7], X_testing.iloc[:,:7], y_training, y_testing)
```

Ex 4: Linear regression model to the training data using only the 7th input variable

```python
linear_regression_r2_score(X_training.iloc[:,6:7], X_testing.iloc[:,6:7], y_training, y_testing)
```

Ridge regression

```python
from sklearn.linear_model import Ridge, Lasso

def ridge_regression_r2_score(X_train, X_test, y_train, y_test, alpha):
  clf_lr = Ridge(alpha=alpha)
  clf_lr.fit(X_train,y_train)
  y_pred = clf_lr.predict(X_test)
  y_pred_train = clf_lr.predict(X_train)
  return {'train':r2_score(y_train,y_pred_train), 
          'test':r2_score(y_test,y_pred), 
          'coeff':clf_lr.coef_,
          'constant': clf_lr.intercept_
          }
```

Ex 5: ridge regression model to the training data using all input variables with alpha=10

```python
ridge_regression_r2_score(X_training, X_testing, y_training, y_testing, alpha=10)
```

Ex 6: ridge regression model to the training data using all input variables with alpha=100


```python
ridge_regression_r2_score(X_training, X_testing, y_training, y_testing, alpha=100)
```

Ex 7: Lasso regression model to the training data using all input variables with alpha=10.

```python
def lasso_regression_r2_score(X_train, X_test, y_train, y_test, alpha):
  clf_lr = Lasso(alpha=alpha)
  clf_lr.fit(X_train,y_train)
  y_pred = clf_lr.predict(X_test)
  y_pred_train = clf_lr.predict(X_train)
  return {'train':r2_score(y_train,y_pred_train), 
          'test':r2_score(y_test,y_pred), 
          'coeff':clf_lr.coef_,
          'constant': clf_lr.intercept_
          }
```

```python
lasso_regression_r2_score(X_training, X_testing, y_training, y_testing, alpha=10)
```
