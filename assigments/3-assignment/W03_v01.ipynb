```python
# Create a new column for click-to-sales-conversion (i.e. sales / click )
# Only look at rows with at least 100 clicks.
df_over_100_clicks = df[df['click'] > 100]
df_over_100_clicks

skus_sales = df_over_100_clicks[['SKU', 'sales','click']]
```

```python
total_sales_per_SKU = {}

for sku in df_over_100_clicks['SKU'].unique():
    total_sales_per_SKU[sku] = skus_sales.loc[skus_sales['SKU'] == sku]['sales'].sum()

total_clicks_per_SKU = {}
for sku in df_over_100_clicks['SKU'].unique():
    total_clicks_per_SKU[sku] = skus_sales.loc[skus_sales['SKU'] == sku]['click'].sum()

total_sales_per_SKU
```

```python
total_clicks_per_SKU['ZG885592']
```

```python
def click_to_sales_conversion(row):
    return round(total_sales_per_SKU[row['SKU']]/total_clicks_per_SKU[row['SKU']],4) 


def test(sku):
    return sku['SKU']
    
#df_over_100_clicks['click-to-sales-conversion'] = df_over_100_clicks.apply(lambda row : click_to_sales_conversion(row['SKU'],
click_sales_conversion = df_over_100_clicks.apply(lambda row : click_to_sales_conversion(row), axis = 1)
```

```python
df['click-to-sales-conversion'] = click_sales_conversion.reindex(list(range(275265)),fill_value=0)
```
